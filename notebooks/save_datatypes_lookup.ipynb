## define, make better

```python
special_objects_len_dict = {
    'bg_fips': 12,
    'census_tract_fips': 11,
    'county_fips': 5,
    'state_fips': 2,
    'state': 2,
    'zip': 5,
    'msa_code': 5
}

special_decimal_dict = {
    'blockgroup_center_lat': 'decimal(8,6)',
    'blockgroup_center_lng': 'decimal(8,6)'
}

not_null_cols = [
    'bg_fips', 'census_tract_fips', 'county_fips', 'state_fips', 'state',
       'county_name', 'blockgroup_center_lat', 'blockgroup_center_lng',
    # these are also not null, though they could be
    # 'population', 'households'
]
```

```python
ddl_dtype_dict_better = ddl_dtype_dict.copy()

for key, val in special_objects_len_dict.items():
    ddl_dtype_dict_better[key] = f'varchar({val})'

for key, val in special_decimal_dict.items():
    ddl_dtype_dict_better[key] = val
```

## define DDL string

### add nulls

```python
ddl_dtype_dict_for_string = ddl_dtype_dict_better.copy()

for col in not_null_cols:
    ddl_dtype_dict_for_string[col] += " not null"
```

```python
ddl_dtype_dict_for_string
```

```python
ddl_create_str = ",\n".join([f"{key}   {val}" for key, val in ddl_dtype_dict_for_string.items()])
```

```python
print(ddl_create_str)
```

```python
with open(f'{ACS_BG_FILENAME}_DDL_string.txt', 'w') as f:
    f.write(ddl_create_str)
```

## add nulls to .csv

```python
ddl_dtype_list_with_nulls = []

for col, dtype in ddl_dtype_dict_better.items():
    data = [col, dtype]
    if col in not_null_cols:
        data.append("not null")
    else:
        data.append("")

    ddl_dtype_list_with_nulls.append(data)
```

```python
ddl_dtype_list_with_nulls[:5]
```

```python
ddl_dtype_df = pd.DataFrame(data=ddl_dtype_list_with_nulls, columns=['column', 'dtype', 'nullable'])
```
